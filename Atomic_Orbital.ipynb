```python
# import libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng()
```

```python
N = 51
r = np.linspace(0, 30, N)
h_r = r[1] - r[0]

psi = r * (1 - r / 6) * np.exp(-r / 3)
prob_r = psi ** 2 * r
# normalise this probability so that the integral over the domain is 1.
rho_r = prob_r / np.sum(prob_r) / h_r
F_r = np.cumsum(rho_r) * h_r

x = np.interp(rng.uniform(size=10000), F_r, r)
h = plt.hist(x, bins=50, density=True)
plt.plot(r, rho_r, 'r')
plt.title('Histogram of random sample together with probability density');
```

# Electron probabilty distribution 

```python
N = 101

r = np.linspace(0, 30, N)
psi = r * (1 - r / 6) * np.exp(-r / 3)
F_r = np.cumsum(psi ** 2 * r)
F_r /= np.max(F_r)

theta = np.linspace(0, 2 * np.pi, N)
F_theta = np.cumsum(np.sin(theta) ** 2)
F_theta /= np.max(F_theta)

M = 5000
r_picture = np.interp(rng.uniform(size=M), F_r, r)
theta_picture = np.interp(rng.uniform(size=M), F_theta, theta)
plt.plot(r_picture * np.sin(theta_picture), r_picture * np.cos(theta_picture),
         '.', markersize=1);
```
